```python
# from collections import Counter

def solution(genres, plays):
    a = dict()
    b = dict()
    n = len(genres)
    for i in range(n):
        if genres[i] not in a:
            a[genres[i]] = 0
        a[genres[i]]+=plays[i]
        if genres[i] not in b:
            b[genres[i]] = [[-1,-1],[-1,-1]]
        b[genres[i]].append([plays[i],i])
        b[genres[i]].sort(key = lambda x : (-x[0],x[1]))
        b[genres[i]].pop()
    result=[]
    for g,p in sorted(a.items(), key = lambda x : -x[1]):
        for k,v in b[g]:
            if k !=-1:
                result.append(v)
    return result
```

```python
def solution(genres, plays):
    answer = []
    d = {e:[] for e in set(genres)}
    for e in zip(genres, plays, range(len(plays))):
        d[e[0]].append([e[1] , e[2]])
    genreSort =sorted(list(d.keys()), key= lambda x: sum( map(lambda y: y[0],d[x])), reverse = True)
    for g in genreSort:
        temp = [e[1] for e in sorted(d[g],key= lambda x: (x[0], -x[1]), reverse = True)]
        answer += temp[:min(len(temp),2)]
    return answer
```
