```python
def solution(phone_book):
    a=set()
    for p in phone_book:
        a.add(len(p))
    b=dict()
    for p in phone_book:
        b[p]=0
    for p in phone_book:
        for len_a in a:
            if len(p) >= len_a:
                c = p[:len_a]
                if c in b.keys():
                    b[c]+=1
    for v in b.values():
        if v > 1 :
            return False
    return True
```

```python
phone_book=["119", "97674223", "1195524421"]
solution(phone_book)
```

```python
def solution(phoneBook):
    phoneBook = sorted(phoneBook)

    for p1, p2 in zip(phoneBook, phoneBook[1:]):
        if p2.startswith(p1):
            return False
    return True
```
