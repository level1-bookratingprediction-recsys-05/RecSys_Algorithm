```python
class Solution:
    def divisorGame(self, n: int) -> bool:
        dp = [-1] * 1001
        dp[1] = (False,0)
        for i in range(2,n+1):
            for m in [i-k for k in [j for j in range(1, int(i**0.5)+1) if i%j==0]]:
                if not dp[m][0]:
                    dp[i] = (True, m)
                    break
            else:
                dp[i] = (False, i-1)
        return dp[n][0]
```
