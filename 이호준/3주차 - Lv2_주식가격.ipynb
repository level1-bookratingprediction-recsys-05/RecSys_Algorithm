```python
def solution(prices):
    answer = list(range(len(prices)-1,-1,-1))
    stack = []
    for i,v in enumerate(prices):
        while stack and stack[-1][1]>v:
            answer[stack.pop()[0]]-=len(prices)-1-i
        stack.append([i,v])
#         print(stack)
    return answer
```

```python
prices = [1, 2, 1, 2, 3]
solution(prices)
```
